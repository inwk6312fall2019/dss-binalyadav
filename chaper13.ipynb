```python
import string
```

```python
f=open("words.txt", "r" ,encoding = "ISO-8859-1")
```

```python
file = f.read()
```

```python
file = file.rstrip("\n\r")
```

```python
file = file.replace('\n',' ')
```

```python
file = file.translate(str.maketrans('', '', string.punctuation))
```

```python
words = file.split()
```

```python
#13.1
wordList = []
for word in words:
    word = word.lower()
    wordList.append(word.strip())
```

```python
f=open("2600-0.txt", "r" ,encoding = "ISO-8859-1")
```

```python
book = f.read()
```

```python
#13.2
book = book.rstrip("\n\r")
book = book.replace('\n',' ')
book = book.translate(str.maketrans('', '', string.punctuation))
words = book.split()
wordList = []
for word in words:
    word = word.lower()
    wordList.append(word.strip())
```

```python
wordList
```

```python
len(wordList)
```

```python
uniqueWords = set(wordList)
```

```python

wordOccurance = {}
for word in uniqueWords:
    occurance = wordList.count(word)
    wordOccurance[word] = occurance
```

```python
wordOccurance
```

```python
#13.3


#  we have to sort wordOccurance Dictionary by values in descending order, first 20 will be the most frequently occuring words
wordOccuranceDesc = sorted(wordOccurance.items(), key=lambda kv: kv[1],reverse=True) #with help of google
```
